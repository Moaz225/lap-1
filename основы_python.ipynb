# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def is_unique(s):
    """
    Проверяет, встречается ли каждый символ в строке только один раз.
    
    :param s: входная строка
    :return: True, если все символы уникальны, иначе False
    """
    # Множество для хранения уникальных символов
    seen = set()
    
    # Проходим по каждому символу строки
    for char in s:
        # Если символ уже встречался, возвращаем False
        if char in seen:
            return False
        # Добавляем символ в множество
        seen.add(char)
    
    # Если не найдено повторяющихся символов, возвращаем True
    return True


# Тесты
def test_is_unique():
    # Тест 1: строка с уникальными символами
    assert is_unique("abcde") == True, "Тест 1 не пройден!"
    
    # Тест 2: строка с повторяющимися символами
    assert is_unique("aabcde") == False, "Тест 2 не пройден!"
    
    # Тест 3: пустая строка
    assert is_unique("") == True, "Тест 3 не пройден!"  # Пустая строка считается как строка с уникальными символами
    
    # Тест 4: строка с одним символом
    assert is_unique("a") == True, "Тест 4 не пройден!"
    
    # Тест 5: строка с повторяющимися символами в разном регистре
    assert is_unique("AbcdeA") == False, "Тест 5 не пройден!"
    
    print("Все тесты пройдены!")

# Запуск тестов
test_is_unique()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_bits(n):
    """
    Возвращает количество бит, равных 1, в двоичном представлении числа n.
    
    :param n: положительное целое число
    :return: количество битов, равных 1
    """
    # Преобразуем число в двоичное представление и считаем количество единиц
    return bin(n).count('1')


# Тесты
def test_count_bits():
    # Тест 1: число 0 (в двоичной системе: 0)
    assert count_bits(0) == 0, "Тест 1 не пройден!"
    
    # Тест 2: число 1 (в двоичной системе: 1)
    assert count_bits(1) == 1, "Тест 2 не пройден!"
    
    # Тест 3: число 5 (в двоичной системе: 101)
    assert count_bits(5) == 2, "Тест 3 не пройден!"
    
    # Тест 4: число 15 (в двоичной системе: 1111)
    assert count_bits(15) == 4, "Тест 4 не пройден!"
    
    # Тест 5: число 1023 (в двоичной системе: 1111111111)
    assert count_bits(1023) == 10, "Тест 5 не пройден!"
    
    print("Все тесты пройдены!")

# Запуск тестов
test_count_bits()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    """
    Возвращает количество итераций умножения цифр числа, чтобы получить однозначное число.
    
    :param n: положительное целое число
    :return: количество итераций
    """
    # Счётчик итераций
    count = 0
    
    # Пока число состоит из более чем одной цифры
    while n >= 10:
        product = 1
        # Преобразуем число в строку и перемножаем все его цифры
        for digit in str(n):
            product *= int(digit)
        # Присваиваем результат умножения
        n = product
        count += 1
    
    return count


# Тесты
def test_magic():
    # Тест 1: число 39
    assert magic(39) == 3, "Тест 1 не пройден!"  # 39 => 27 => 14 => 4
    
    # Тест 2: число 4 (уже однозначное число)
    assert magic(4) == 0, "Тест 2 не пройден!"  # Число уже однозначное
    
    # Тест 3: число 999
    assert magic(999) == 4, "Тест 3 не пройден!"  # 999 => 729 => 126 => 12 => 2
    
    # Тест 4: число 25
    assert magic(25) == 2, "Тест 4 не пройден!"  # 25 => 10 => 0
    
    # Тест 5: число 1 (уже однозначное)
    assert magic(1) == 0, "Тест 5 не пройден!"  # Число уже однозначное
    
    print("Все тесты пройдены!")

# Запуск тестов
test_magic()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
import numpy as np

def mse(pred, true):
    """
    Вычисляет среднеквадратическое отклонение между двумя векторами.
    
    :param pred: первый вектор (предсказанные значения)
    :param true: второй вектор (истинные значения)
    :return: среднеквадратическое отклонение (float)
    """
    # Проверяем, что длины векторов совпадают
    if len(pred) != len(true):
        raise ValueError("Векторы должны быть одинаковой длины")
    
    # Если векторы пустые, возвращаем 0 или выбрасываем исключение
    if len(pred) == 0:
        return 0.0
    
    # Вычисляем среднеквадратическое отклонение
    return np.mean((np.array(pred) - np.array(true)) ** 2)


# Тесты
def test_mse():
    # Тест 1: Одинаковые векторы
    assert mse([1, 2, 3], [1, 2, 3]) == 0.0, "Тест 1 не пройден!"  # Ошибка должна быть 0
    
    # Тест 2: Векторы с разными значениями
    assert np.isclose(mse([1, 2, 3], [4, 5, 6]), 9.0), "Тест 2 не пройден!"  # Должно быть 9.0
    
    # Тест 3: Векторы с небольшими различиями
    assert np.isclose(mse([1, 1, 1], [1, 2, 1]), 0.3333333333333333), "Тест 3 не пройден!"
    
    # Тест 4: Векторы нулевой длины (пустые)
    assert mse([], []) == 0.0, "Тест 4 не пройден!"  # Ошибка должна быть 0 для пустых векторов
    
    # Тест 5: Векторы разной длины
    try:
        mse([1, 2], [1, 2, 3])
        assert False, "Тест 5 не пройден!"  # Ожидается ошибка
    except ValueError:
        assert True, "Тест 5 пройден!"
    
    print("Все тесты пройдены!")

# Запуск тестов
test_mse()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
from collections import Counter

def prime_factors(n):
    """
    Возвращает строковое представление разложения числа на простые множители.
    
    :param n: целое положительное число
    :return: строка вида "(n1p1)(n2p2)…(nkpk)"
    """
    i = 2
    factors = []
    # Разделяем n на простые множители
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    
    # Подсчет кратностей каждого простого множителя
    factor_counts = Counter(factors)
    
    # Формируем строку результата
    result = ""
    for factor, count in sorted(factor_counts.items()):
        if count == 1:
            result += f"({factor})"
        else:
            result += f"({factor}**{count})"
    
    return result


# Тесты
def test_prime_factors():
    # Тест 1: число 86240
    assert prime_factors(86240) == "(2**5)(5)(7**2)(11)", "Тест 1 не пройден!"
    
    # Тест 2: простое число 13
    assert prime_factors(13) == "(13)", "Тест 2 не пройден!"
    
    # Тест 3: число 60 (разложение: 2^2 * 3 * 5)
    assert prime_factors(60) == "(2**2)(3)(5)", "Тест 3 не пройден!"
    
    # Тест 4: число 100 (разложение: 2^2 * 5^2)
    assert prime_factors(100) == "(2**2)(5**2)", "Тест 4 не пройден!"
    
    # Тест 5: число 1 (разложения нет, возвращаем пустую строку)
    assert prime_factors(1) == "", "Тест 5 не пройден!"
    
    print("Все тесты пройдены!")

# Запуск тестов
test_prime_factors()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    """
    Определяет, можно ли из данного количества кубиков построить пирамиду.
    Возвращает количество уровней пирамиды k, если это возможно, иначе строку "It is impossible".
    
    :param number: целое число, задающее количество кубиков
    :return: количество уровней k или строка "It is impossible"
    """
    k = 0
    total_cubes = 0
    
    # Добавляем квадраты чисел, пока общая сумма не превысит number
    while total_cubes < number:
        k += 1
        total_cubes += k ** 2
    
    # Если сумма равна числу, возвращаем k
    if total_cubes == number:
        return k
    else:
        return "It is impossible"


# Тесты
def test_pyramid():
    # Тест 1: число 14 (разложение 1^2 + 2^2 + 3^2)
    assert pyramid(14) == 3, "Тест 1 не пройден!"  # 14 = 1 + 4 + 9
    
    # Тест 2: число 30 (разложение 1^2 + 2^2 + 3^2 + 4^2)
    assert pyramid(30) == 4, "Тест 2 не пройден!"  # 30 = 1 + 4 + 9 + 16
    
    # Тест 3: число 5 (разложение 1^2 + 2^2)
    assert pyramid(5) == 2, "Тест 3 не пройден!"  # 5 = 1 + 4
    
    # Тест 4: число 1 (разложение 1^2)
    assert pyramid(1) == 1, "Тест 4 не пройден!"  # 1 = 1
    
    # Тест 5: число 0 (невозможно построить пирамиду)
    assert pyramid(0) == 0, "Тест 5 не пройден!"  # 0 кубиков — это 0 уровней
    
    print("Все тесты пройдены!")

# Запуск тестов
test_pyramid()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(n):
    """
    Определяет, является ли число сбалансированным.
    
    :param n: Положительное целое число
    :return: True, если число сбалансированное, иначе False
    """
    # Преобразуем число в строку, чтобы легко получить цифры
    str_n = str(n)
    length = len(str_n)
    
    # Если количество цифр четное
    if length % 2 == 0:
        mid = length // 2
        left_part = str_n[:mid]
        right_part = str_n[mid:]
    # Если количество цифр нечетное
    else:
        mid = length // 2
        left_part = str_n[:mid]
        right_part = str_n[mid+1:]
    
    # Считаем суммы цифр до и после средней/средних
    sum_left = sum(int(digit) for digit in left_part)
    sum_right = sum(int(digit) for digit in right_part)
    
    # Проверяем, равны ли суммы
    return sum_left == sum_right


# Тесты
def test_is_balanced():
    # Тест 1: Сбалансированное число с четным количеством цифр
    assert is_balanced(123321) == True, "Тест 1 не пройден!"
    
    # Тест 2: Несбалансированное число с четным количеством цифр
    assert is_balanced(123421) == False, "Тест 2 не пройден!"
    
    # Тест 3: Сбалансированное число с нечетным количеством цифр
    assert is_balanced(12321) == True, "Тест 3 не пройден!"
    
    # Тест 4: Несбалансированное число с нечетным количеством цифр
    assert is_balanced(12331) == False, "Тест 4 не пройден!"
    
    # Тест 5: Однозначное число (считается сбалансированным)
    assert is_balanced(1) == True, "Тест 5 не пройден!"
    
    print("Все тесты пройдены!")

# Запуск тестов
test_is_balanced()
```
